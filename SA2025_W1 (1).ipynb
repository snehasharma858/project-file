### 1. What is name of car that has the highest horsepower?  

```python


df = pd.read_csv("cars.csv")
df = df.dropna(subset=['horsepower'])


max_hp_row = df[df['horsepower'] == df['horsepower'].max()]


print("Car with the highest horsepower:")
print(max_hp_row[['name', 'horsepower']])
```

### 2. How many cars have mpg  35?

```python
high_mpg_cars = df[df['mpg'] >= 35]
count = len(high_mpg_cars)

print(f"Number of cars with mpg  35: {count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python


# Load the built-in 'mpg' dataset
df = sns.load_dataset("mpg")

# Drop rows with missing horsepower or weight
df = df.dropna(subset=['horsepower', 'weight'])

# Filter cars with horsepower > 100 and weight < 3000
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_cars['origin'].mode().iloc[0]

print(f"The most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")

```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python


df = sns.load_dataset("mpg")
df = df.dropna(subset=['origin', 'acceleration'])

# Filter cars from Japan
japan_cars = df[df['origin'] == 'japan']

# Calculate mean acceleration, rounded to 2 decimal places
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print(f"Mean acceleration of cars from Japan: {mean_acceleration}")

```

### 5. Which year had the highest average mpg?

```python


# Load the dataset
df = sns.load_dataset("mpg")

# Drop rows with missing 'mpg' or 'model_year'
df = df.dropna(subset=['mpg', 'model_year'])


avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()

# Find the year with the highest average mpg
best_year = avg_mpg_by_year.idxmax()
best_avg = round(avg_mpg_by_year.max(), 2)

print(f"The year with the highest average mpg is {best_year} with an average of {best_avg} mpg.")

```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
df = sns.load_dataset("mpg")

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['horsepower', 'weight', 'mpg', 'name'])

# Step 1: Calculate median mpg
median_mpg = df['mpg'].median()

# Step 2: Filter cars with mpg > median
efficient_cars = df[df['mpg'] > median_mpg]

# Step 3: Compute horsepower-to-weight ratio
efficient_cars['hp_weight_ratio'] = efficient_cars['horsepower'] / efficient_cars['weight']

# Step 4: Find max ratio
max_ratio = efficient_cars['hp_weight_ratio'].max()

# Step 5: Filter car(s) with that ratio
best_cars = efficient_cars[efficient_cars['hp_weight_ratio'] == max_ratio]

# Show the results
print("Car(s) with the best horsepower-to-weight ratio (among those with above-median mpg):")
print(best_cars[['name', 'mpg', 'horsepower', 'weight', 'hp_weight_ratio']])

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python

# Load the 'mpg' dataset
df = sns.load_dataset("mpg")

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['mpg', 'model_year', 'origin'])

# Group by 'model_year' and 'origin' to compute average mpg
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set the style
sns.set(style="whitegrid")

# Create the multi-line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

# Customize the plot
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(avg_mpg['model_year'].unique())  # show all model years on x-axis
plt.legend(title='Origin')
plt.tight_layout()

# Show the plot
plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python


# Load the dataset
df = sns.load_dataset("mpg")

# Drop rows with missing values
df = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])

# Set Seaborn style
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    hue_order=['japan', 'europe', 'usa'],
    alpha=0.7,
    palette='Set2'
)

# Titles and labels
plt.title('Horsepower vs. Weight of Cars\n(Size = MPG, Color = Origin)', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()

# Show the plot
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python

df = sns.load_dataset("mpg")

# Drop rows with missing data in relevant columns
df = df.dropna(subset=['name', 'mpg', 'model_year'])

# Group by car 'name' and count unique years
car_years = df.groupby('name')['model_year'].nunique()

# Filter names that appear in more than one year
multi_year_models = car_years[car_years > 1].index

# Filter dataset to include only these models
multi_year_df = df[df['name'].isin(multi_year_models)]

# Group again by 'name' to compute stats
grouped = multi_year_df.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    count=('mpg', 'count'),
    mpg_mean=('mpg', 'mean')
).reset_index()

# Filter for consistent models: std < 1.0
consistent_models = grouped[grouped['mpg_std'] < 1.0]

# Bonus: Sort by count (descending), then mpg_mean (descending)
consistent_models_sorted = consistent_models.sort_values(by=['count', 'mpg_mean'], ascending=[False, False])

# Final result
print("Consistent car models (std(mpg) < 1.0 and > 1 year):\n")
print(consistent_models_sorted[['name', 'count', 'mpg_mean']])

```
